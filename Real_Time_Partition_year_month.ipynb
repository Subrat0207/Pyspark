'''
most of the Traditional DBMS databases will be having Default Date Format is DD-MM-YYYY .  But Cloud Data storage(Spark delta Lake/Databricks tables) will be using YYYY-MM-DD Format.
Here i covered how to convert dd-MM-yyyy format to yyyy-MM-dd format using to_date() function in pyspark
'''

```
emp_csv_df = spark.read.csv('dbfs:/FileStore/tables/Target/Emp.csv',header= True,inferSchema=True)
display(emp_csv_df)
```

```
emp_csv_df.printSchema()
```

```
spark.conf.set('spark.sql.legacy.timeParserPolicy', 'legacy')

```

```
emp_csv_df.show()
```

```
from pyspark.sql.functions import to_date
emp_csv_df = emp_csv_df.withColumn('HIREDATE',to_date('HIREDATE','dd-MM-yyyy')).fillna({'HIREDATE':'9999-09-09'})
display(emp_csv_df)
```

```
emp_csv_df = emp_csv_df.withColumn('YEAR',date_format('HIREDATE','yyyy')).withColumn('MONTH',date_format('HIREDATE','MM'))
emp_csv_df.show()
```

```
dbutils.fs.mkdirs('dbfs:/FileStore/tables/Target/emp_part_month_day_Real_time_issue/')
```

```
%fs ls dbfs:/FileStore/tables/Target/emp_part_month_day_Real_time_issue/
```

```
emp_csv_df.write.format('delta').mode('append').partitionBy('YEAR','MONTH').save('dbfs:/FileStore/tables/Target/emp_part_month_day_Real_time_issue/')
```
